## Deep Dive: Human-Resonant AI Writing Engine

This notebook demonstrates:
1. Multi-tenant HRV profile loading
2. Human-Resonant article generation
3. HRF feedback simulation per paragraph
4. Visualization of HRV across paragraphs
5. Optional HR-PPO reinforcement learning


```python
# Ensure the package is importable
import sys
sys.path.append('..')

import numpy as np
import matplotlib.pyplot as plt
from resonance_os.generation.human_resonant_writer import HumanResonantWriter
from resonance_os.profiles.multi_tenant_hr_profiles import HRVProfileManager
from resonance_os.evolution.hr_rl_trainer import HRWritingEnv, train_hr_ppo

print('Modules imported successfully')
```

## 1️⃣ Load Multi-Tenant HRV Profiles

```python
# Setup profile manager
profile_manager = HRVProfileManager('./profiles/hr_profiles')
tenant = 'default'
profile_name = 'brand_identity_v1'

# Load HRV target vector
target_hrv = profile_manager.load_profile(tenant, profile_name)
print(f"Loaded HRV profile for tenant '{tenant}' profile '{profile_name}': {target_hrv}")
```

## 2️⃣ Initialize Human-Resonant Writer

```python
writer = HumanResonantWriter()
print('Human-Resonant Writer initialized')
```

## 3️⃣ Generate Human-Resonant Article

```python
prompt = 'Write a futuristic AI article integrating multi-agent resonance'
article = writer.generate(prompt)
paragraphs = [p.strip() for p in article.split('.') if p.strip()]

print('=== Generated Article ===')
for i, p in enumerate(paragraphs, 1):
    print(f'Paragraph {i}: {p}')
```

## 4️⃣ Simulate HRF Feedback per Paragraph

```python
# Simulate HRF feedback (how 'human-like' each paragraph is)
hrv_feedback = np.random.rand(len(paragraphs), len(target_hrv))

print('=== Simulated HRV Feedback ===')
print(hrv_feedback)
```

## 5️⃣ Visualize HRV Feedback Across Paragraphs

```python
plt.figure(figsize=(12,6))
for i in range(len(target_hrv)):
    plt.plot(hrv_feedback[:, i], label=f'HRV Dim {i}')
plt.title('Human-Resonant Feedback per Paragraph')
plt.xlabel('Paragraph Index')
plt.ylabel('HRV Score')
plt.legend()
plt.show()
```

## 6️⃣ Optional: Train HR-PPO for Resonance Optimization

```python
env = HRWritingEnv(hrv_dim=len(target_hrv))
model = train_hr_ppo(env, timesteps=2000)
print('✅ HR-PPO model trained for resonance optimization')
```
